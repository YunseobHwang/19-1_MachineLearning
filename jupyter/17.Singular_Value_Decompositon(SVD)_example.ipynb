```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
A = np.matrix([[4, 4], [-3, 3]])

D, V = np.linalg.eig(A.T*A)
print('V:\n', V)
print('D:\n', D)
```

```python
A = np.matrix([[4, 4], [-3, 3]])

D, U = np.linalg.eig(A*A.T)
print('U:\n', U)
print('D:\n', D)
```

```python
U, S, V = np.linalg.svd(A)
print('U:\n', U)
print('S:\n', S)
print('V:\n', V)

s = np.zeros([2, 2])
s[0, 0], s[1, 1] = S[0], S[1]

print('A:\n', A)
print('A:\n', U*s*V.T)
```

```python
X, Y = np.meshgrid(np.arange(0, 3/2*np.pi, np.pi/30), 
                   np.arange(0, 3/2*np.pi, np.pi/30))
Z = np.sin(X*Y*np.pi/3)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1)
plt.show()
```

```python
maxNoise = .1;    # Change this parameter for the amount of noise to add

nRows, nCols = Z.shape
noise = 2*maxNoise*np.random.rand(nRows, nCols)
Znoise = Z + noise
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Znoise, rstride=1, cstride=1)
plt.show()
```

```python
U, S, V =np.linalg.svd(Znoise)
V = V.T
```

```python
plt.figure(figsize=(10, 6))
plt.stem(S)
plt.show()
```

```python
k = 8
low_U = np.asmatrix(U[:,:k])
low_S = np.asmatrix(np.diag(S[:k]))
low_V = np.asmatrix(V[:,:k])
Znoiseless = low_U*low_S*low_V.T

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Znoiseless, rstride=1, cstride=1)
plt.show()
```
