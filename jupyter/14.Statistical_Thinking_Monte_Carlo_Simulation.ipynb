```python
import numpy  as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
n_trials = 100 #sample
n_H = 0

for i in range(n_trials):
    flip = np.random.randint(2)
    if flip == 1:
        n_H +=1
        
print(n_H/n_trials) # sample mean
```

```python
# Expectaion

val  = 0 

for n in range(1,20):
    val += n*(1/2)**n     # 공식
    
print(val)
```

```python
n_trials = 1000

NUM = []

for i in range(n_trials):
    num = 1
    while np.random.randint(2) != 0:   # 0이 아니면 증가 
        num += 1    
    NUM.append(num)
    
print(NUM)
print(np.mean(NUM))
```

```python
# Integration
# by summing up

dx = 0.001
x = np.arange(0, 1, dx)

A = 0
for xk in x:
    A += (xk**2)*dx
    
print(A)
```

```python
# shortened version

A = np.sum(x**2)*dx
print(A)
```

```python
# Monte Carlo Simulation
# pi 계산하기 

n = 5000

y = 2*np.random.rand(n, 1) - 1
x = 2*np.random.rand(n, 1) - 1

idx = np.empty((n,1))
count = 0 

for i in range(n):
    if np.sqrt(x[i]**2 + y[i]**2) < 1:
        count += 1
        idx[i] = 1
    else:
        idx[i] = 0

print((count/n)*4)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x[idx == 0], y[idx == 0], 'b.')
plt.plot(x[idx == 1], y[idx == 1], 'r.')
plt.axis('equal')
plt.show()
```
